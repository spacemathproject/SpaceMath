```python
from spacemathpy import *
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mhiggs = np.linspace(10,1000,1000)
ghttSM = mt['value']/(SMvev['value'])
ghbbSM = mb['value']/(SMvev['value'])
ghtautauSM = mtau['value']/(SMvev['value'])

Whtt = WidthHff(ghttSM,1,mt['value'],mt['value'],mhiggs)
Whtt[np.isnan(Whtt)] = 0
Whww = WidthHWW(gw['value']*mW['value'],mhiggs)
Whww[np.isnan(Whww)] = 0
Whzz = WidthHZZ(gz['value']*mZ['value'],mhiggs)
Whzz[np.isnan(Whzz)] = 0

totw = (WidthHgg(ghttSM,ghbbSM,mhiggs) + 
       WidthHff(ghtautauSM,1,mtau['value'],mtau['value'],mhiggs) + 
       Whtt + 
       WidthHff(ghbbSM,1,mb['value'],mb['value'],mhiggs) + 
       Whww + 
       Whzz)


plt.semilogy(mhiggs,WidthHgg(ghttSM,ghbbSM,mhiggs)/totw,'.',label='g')
plt.semilogy(mhiggs,WidthHff(ghtautauSM,1,mtau['value'],mtau['value'],mhiggs)/totw,'.',label=r'$\tau$')
plt.semilogy(mhiggs,WidthHff(ghttSM,1,mt['value'],mt['value'],mhiggs)/totw,'.',label='t')
plt.semilogy(mhiggs,WidthHff(ghbbSM,1,mb['value'],mb['value'],mhiggs)/totw,'.',label='b')
plt.semilogy(mhiggs,WidthHWW(gw['value']*mW['value'],mhiggs)/totw,'.',label='W')
plt.semilogy(mhiggs,WidthHZZ(gz['value']*mZ['value'],mhiggs)/totw,'.',label='Z')

plt.grid()
plt.legend()
```

```python
plt.semilogy(mhiggs,WidthHgg(ghttSM,ghbbSM,mhiggs),'.',label='g')
plt.semilogy(mhiggs,WidthHff(ghtautauSM,1,mtau['value'],mtau['value'],mhiggs),'.',label=r'$\tau$')
plt.semilogy(mhiggs,WidthHff(ghttSM,1,mt['value'],mt['value'],mhiggs),'.',label='t')
plt.semilogy(mhiggs,WidthHff(ghbbSM,1,mb['value'],mb['value'],mhiggs),'.',label='b')
plt.semilogy(mhiggs,WidthHWW(gw['value']*mW['value'],mhiggs),'.',label='W')
plt.semilogy(mhiggs,WidthHZZ(gz['value']*mZ['value'],mhiggs),'.',label='Z')
plt.semilogy(mhiggs,totw,'.',label='Tot')
plt.grid()
plt.legend(loc='lower right')
```
